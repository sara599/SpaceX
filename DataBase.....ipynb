 ### **❓Launch Success Rate**

```python
# Overall success rate of SpaceX launches:
query = """
SELECT 
    COUNT(*) AS total_launches,
    SUM(success) AS successful_launches,
    (SUM(success) / COUNT(*)) * 100 AS success_rate
FROM cleaned_data;
"""

try:
    cursor = connection.cursor()
    cursor.execute(query)  # Execute the SQL query
    result = cursor.fetchone()  # Fetch the result
    print(f"Total Launches: {result[0]}, Successful Launches: {result[1]}, Success Rate: {result[2]:.2f}%")
except Error as e:
    print(f"Error executing query: {e}")
```

```python
# Success rate by rocket type:
query = """
SELECT 
    rocket,
    COUNT(*) AS total_launches,
    SUM(success) AS successful_launches,
    (SUM(success) / COUNT(*)) * 100 AS success_rate
FROM cleaned_data
GROUP BY rocket;
"""

try:
    cursor = connection.cursor()
    cursor.execute(query)  # Execute the SQL query
    results = cursor.fetchall()  # Fetch all results

    for row in results:
        rocket, total_launches, successful_launches, success_rate = row
        print(f"Rocket: {rocket}, Total Launches: {total_launches}, Successful Launches: {successful_launches}, Success Rate: {success_rate:.2f}%")
except Error as e:
    print(f"Error executing query: {e}")
```

 ### **❓Payload Analysis**

```python
# SQL query to find the most common types of payloads launched
query = """
SELECT 
    name_payload, 
    COUNT(*) AS launch_count
FROM cleaned_data
GROUP BY name_payload
ORDER BY launch_count DESC;
"""

try:
    cursor = connection.cursor()  # Create a cursor object
    cursor.execute(query)  # Execute the SQL query
    results = cursor.fetchall()  # Fetch all rows from the query result
    
    # Print the results or process them further
    print("Most Common Types of Payloads Launched:")
    for row in results:
        print(f"Payload Name: {row[0]}, Launch Count: {row[1]}")
    
except Error as e:
    print(f"Error executing query: {e}")
```

```python
# Mass of payloads affecting success rate:

# SQL query to analyze launches based on payload mass
query = """
SELECT 
    mass_kg, 
    COUNT(*) AS launch_count,
    SUM(success) AS successful_launches,
    (SUM(success) / NULLIF(COUNT(*), 0)) * 100 AS success_rate
FROM cleaned_data
GROUP BY mass_kg
ORDER BY mass_kg;
"""

try:
    cursor = connection.cursor()  # Create a cursor object
    cursor.execute(query)  # Execute the SQL query
    results = cursor.fetchall()  # Fetch all rows from the query result
    
    # Print the results or process them further
    print("Launch Analysis Based on Payload Mass:")
    for row in results:
        mass = row[0] if row[0] is not None else "N/A"  # Handle potential NULL mass values
        launch_count = row[1]
        successful_launches = row[2]
        success_rate = row[3] if row[3] is not None else 0.0  # Handle potential NULL success rates
        
        print(f"Mass (kg): {mass}, Launch Count: {launch_count}, Successful Launches: {successful_launches}, Success Rate: {success_rate:.2f}%")
    
except Error as e:
    print(f"Error executing query: {e}")
```

 ### **❓Temporal Analysis**

```python
# Number of launches by year:
# SQL query to count launches per year
query = """
SELECT 
    YEAR(date_utc) AS launch_year, 
    COUNT(*) AS launch_count
FROM cleaned_data
GROUP BY launch_year
ORDER BY launch_year;
"""

try:
    # Create a cursor object
    cursor = connection.cursor()  
    # Execute the SQL query
    cursor.execute(query)  
    # Fetch all rows from the query result
    results = cursor.fetchall()  
    
    # Print the results or process them further
    print("Launch Count by Year:")
    for row in results:
        launch_year = row[0]
        launch_count = row[1]
        print(f"Year: {launch_year}, Launch Count: {launch_count}")
    
except Error as e:
    # Print error message if query execution fails
    print(f"Error executing query: {e}")
finally:
    # Ensure the cursor is closed after execution
    if cursor:
        cursor.close()  
```

```python
# Launches per month:

# SQL query to count launches per month
query = """
SELECT 
    MONTH(date_utc) AS launch_month,
    COUNT(*) AS launch_count
FROM cleaned_data
GROUP BY launch_month
ORDER BY launch_month;
"""

try:
    # Create a cursor object
    cursor = connection.cursor()  
    # Execute the SQL query
    cursor.execute(query)  
    # Fetch all rows from the query result
    results = cursor.fetchall()  
    
    # Print the results or process them further
    print("Launch Count by Month:")
    for row in results:
        launch_month = row[0]  # Month (1-12)
        launch_count = row[1]   # Count of launches
        print(f"Month: {launch_month}, Launch Count: {launch_count}")
    
except Error as e:
    # Print error message if query execution fails
    print(f"Error executing query: {e}")
finally:
    # Ensure the cursor is closed after execution
    if cursor:
        cursor.close()  
```

 ### **❓Data Completeness and Quality**

```python
# Checking for unknown values in key columns:

# SQL query to count records and unknown columns
query = """
SELECT 
    COUNT(*) AS total_records,
    COUNT(CASE WHEN name_payload = 'Unknown' THEN 1 END) AS unknown_name_payload,
    COUNT(CASE WHEN mass_kg IS NULL THEN 1 END) AS missing_mass_kg,
    COUNT(CASE WHEN type = 'Unknown' THEN 1 END) AS unknown_type,
    COUNT(CASE WHEN orbit = 'Unknown' THEN 1 END) AS unknown_orbit,
    COUNT(CASE WHEN customers= 'Unknown' THEN 1 END) AS unknown_customers,
    COUNT(CASE WHEN payloads = 'Have_no_payloads' THEN 1 END) AS unknown_payloads
FROM cleaned_data;
"""

try:
    # Create a cursor object
    cursor = connection.cursor()  
    # Execute the SQL query
    cursor.execute(query)  
    # Fetch the result
    result = cursor.fetchone()  
    
    # Print the results
    total_records = result[0]  # Total records count
    unknown_name_payload = result[1]  # Count of unknown name_payload
    missing_mass_kg = result[2]  # Count of missing mass_kg
    missing_type = result[3]  # Count of missing type
    missing_orbit = result[4]  # Count of missing orbit
    missing_customers = result[5]  # Count of missing customers
    unknown_payloads = result[6]  # Count of unknown payloads

    print("Record Counts in cleaned_data:")
    print(f"Total Records: {total_records}")
    print(f"Unknown Name Payload: {unknown_name_payload}")
    print(f"Missing Mass (kg): {missing_mass_kg}")
    print(f"Missing Type: {missing_type}")
    print(f"Missing Orbit: {missing_orbit}")
    print(f"Missing Customers: {missing_customers}")
    print(f"Unknown Payloads: {unknown_payloads}")
    
except Error as e:
    # Print error message if query execution fails
    print(f"Error executing query: {e}")
finally:
    # Ensure the cursor is closed after execution
    if cursor:
        cursor.close()  
```
