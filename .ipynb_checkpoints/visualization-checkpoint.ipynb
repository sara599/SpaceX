```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("cleaned_data.csv")
```

#### <font color='#FF5893'>What factors have the most significant impact on the success of launches?</font>

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
sns.countplot(data=df, x='rocket', hue='success')
plt.title('Launch Success by Rocket Type')

# Right-align the x-axis labels
plt.xticks(rotation=45, ha='right')

plt.show()
```

#### <font color='#FF5893'> What are the most common payload types being launched, and which have the highest success rate?</font>

```python
# Payload types distribution
df['type'].value_counts().plot(kind='bar')
plt.title('Distribution of Payload Types')
plt.show()

# Success by payload type
sns.countplot(data=df, x='type', hue='success')
plt.title('Launch Success by Payload Type')
plt.xticks(rotation=45, ha='right')

plt.show()
plt.show()
```

####### <font color='#FF5893'> Are certain launchpads more successful than others?</font>

```python
sns.countplot(data=df, x='launchpad', hue='success')
plt.title('Launch Success by Launchpad')
plt.xticks(rotation=45, ha='right')
plt.show()
```

#### <font color='#FF5893'>  How does the mass of a payload correlate with the success of the launch and orbit type?</font>

```python
sns.scatterplot(data=df, x='mass_kg', y='success', hue='orbit')
plt.title('Payload Mass vs Launch Success and Orbit Type')
plt.show()
```

#### <font color='#FF5893'> Are there trends in launch success over time?</font>

```python
df['date_utc'] = pd.to_datetime(df['date_utc'])
df['year'] = df['date_utc'].dt.year

# Line plot of success over time
df.groupby('year')['success'].mean().plot(kind='line')
plt.title('Launch Success Over Time')
plt.show()
```

#### <font color='#FF5893'>  Which customers tend to use specific rockets or launchpads more frequently?</font>
How do the top 10 customers distribute their usage across different rockets?

```python
# Get the top 10 customers by the number of launches
top_10_customers = df['customers'].value_counts().nlargest(10).index

# Filter the dataset for only the top 10 customers
df_top_10 = df[df['customers'].isin(top_10_customers)]

# Plot rocket usage for the top 10 customers
sns.countplot(data=df_top_10, x='rocket', hue='customers')
plt.title('Rocket Usage by Top 10 Customers')
plt.xticks(rotation=45, ha='right')
plt.show()
```

#### <font color='#FF5893'>How often are launches conducted, and are there any patterns in the frequency of launches?</font>

```python
# Time series of launch frequency
df['date_utc'] = pd.to_datetime(df['date_utc'])
df.set_index('date_utc').resample('M')['name_launch'].count().plot(kind='line')
plt.title('Monthly Launch Frequency Over Time')
plt.ylabel('Number of Launches')
plt.show()
```

#### <font color='#FF5893'> Which orbits are most frequently targeted by launches, and how successful are launches to different orbits</font>

```python
# Number of launches by orbit
sns.countplot(data=df, x='orbit', hue='success')
plt.title('Launches by Orbit and Success Rate')
plt.xticks(rotation=45)
plt.show()
```

#### <font color='#FF5893'>  Is there any relationship between the rocket type and the payload mass or type?</font>

```python
# Boxplot of payload mass by rocket type
sns.boxplot(data=df, x='rocket', y='mass_kg')
plt.title('Payload Mass Distribution by Rocket Type')
plt.xticks(rotation=45)
plt.show()
```


#### <font color='#FF5893'> Which customers have the highest success rates for launches?</font>
What is the success rate of the top 10 customers, and which customers are more likely to have successful launches?

```python
# Success rate by customer
# Get the top 10 customers by the number of launches
top_10_customers = df['customers'].value_counts().nlargest(10).index

# Filter the dataset for only the top 10 customers
df_top_10 = df[df['customers'].isin(top_10_customers)]

# Calculate the success rate for the top 10 customers
customer_success = df_top_10.groupby('customers')['success'].mean().reset_index()

# Plot success rate by customer
sns.barplot(data=customer_success, x='customers', y='success')
plt.title('Success Rate by Top 10 Customers')
plt.xticks(rotation=90)
plt.show()
```

 #### <font color='#FF5893'> Is there a correlation between payload mass and the likelihood of a successful launch?</font>


```python
# Scatter plot of payload mass and success
sns.scatterplot(data=df, x='mass_kg', y='success')
plt.title('Payload Mass vs Launch Success')
plt.show()
```

 #### <font color='#FF5893'>What factors contribute to failed launches? Are there patterns in failure based on rocket type, launchpad, or payload mass?</font>



```python
# Correlation heatmap for failure analysis
failure_data = df[df['success'] == 0]  # Only look at failed launches
sns.heatmap(failure_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation of Features in Failed Launches')
plt.show()
```

  #### <font color='#FF5893'> How frequently is each launchpad used, and what is the success rate per launchpad?
</font>

```python
# Launch frequency by launchpad
launchpad_freq = df['launchpad'].value_counts().reset_index()
sns.barplot(data=launchpad_freq, x='index', y='launchpad')
plt.title('Launch Frequency by Launchpad')
plt.xticks(rotation=45)
plt.show()
```

  #### <font color='#FF5893'>Are there any notable patterns in upcoming launches? Which rockets, launchpads, and orbits are planned for future launches?

</font>

```python
# Visualizing upcoming launches by rocket type
sns.countplot(data=df[df['upcoming'] == True], x='rocket')
plt.title('Upcoming Launches by Rocket Type')
plt.xticks(rotation=45)
plt.show()

```

  #### <font color='#FF5893'>How has the usage of different rocket types evolved over time?


</font>

```python
# Line plot of rocket usage over time
rocket_time_data = df.groupby([df['date_utc'].dt.year, 'rocket']).size().unstack().fillna(0)
rocket_time_data.plot(kind='line', stacked=True)
plt.title('Rocket Usage Over Time')
plt.ylabel('Number of Launches')
plt.show()
```
